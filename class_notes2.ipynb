```python
s = "Python"

# Access first and last character
print(s[0])
print(s[-1])

# Slicing: start:end (end not included)
print(s[0:4])  # Pyth
print(s[2:])   # thon
print(s[:3])   # Pyt

# Step in slicing
print(s[::2])  # Picks every second character
print(s[::-1]) # Reverse the string
```

```python
s = "Python"
s = 'J' + s[1:]
print(s)
```

```python


s = s + " is fun"
print(s)

# Repetition
print("z" * 5)

s = "Hello World"

```

```python
# Case conversion
print(s.upper())
print(s.lower())

# Splitting
print(s.split())      # Split by space
print(s.split('o'))   # Split by character 'o'

# Count and find
print(s.count('l'))   # Number of 'l'
print(s.find('o'))    # Index of first 'o'

# Boolean checks
print(s.isalpha())
print(s.isalnum())
print(s.islower())
print(s.isupper())
print(s.isspace())
print(s.istitle())

# Endswith check
print(s.endswith("World"))
```

Lists

Lists are collections of items that can be of mixed types. Unlike strings, lists are mutable, so you can modify them.

Creating Lists


```python
numbers = [1, 2, 3]
mixed = ["Python", 3.14, True]

# Empty list
empty = []
#List Length
print(len(numbers))

#Indexing and Slicing Lists
my_list = ['a', 'b', 'c', 1, 2]

# Access single element
print(my_list[0])

# Slice
print(my_list[1:])
print(my_list[:4:2])  # Every second element

```

```python
#Concatenation and Repetition
lst = [1, 2, 3]
# Concatenate lists
print(lst + [4, 5])

# Permanent update
lst = lst + [6]

# Repeat elements
print(lst * 2)
```

List Methods

```python
l = [1, 2, 3]

# Append at the end
l.append(4)

# Pop last element
item = l.pop()
print(item)

# Remove first occurrence
l.remove(2)

# Insert at index
l.insert(1, 99)

# Count occurrences
print(l.count(3))

# Find index of element
print(l.index(3))

# Sort and reverse
l.reverse()
l.sort()
```

Nested Lists

Lists can contain other lists, forming matrices or complex structures:


```python

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access elements
print(matrix[0][1])  # 2
print(matrix[2][0])  # 7

```

List Comprehensions

Quick way to build lists:


```python
# Get first element from each sublist
first_col = [row[0] for row in matrix]
print(first_col)

# Equivalent using a for loop
first_col = []
for row in matrix:
    first_col.append(row[0])
print(first_col)
```
