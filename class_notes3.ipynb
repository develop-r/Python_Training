Creating Tuples

Tuples are created using parentheses () and separating elements with commas.



```python
# Basic tuple
t1 = (1, 2, 3)

# Mixed data types
t2 = ("python", 100, True)

# Single element tuple (comma required)
t3 = (5,)

#Accessing Tuple Elements

#Indexing and slicing work just like lists.

t = ("apple", "banana", "cherry")

print(t[0])
print(t[-1])
print(t[0:2])
#T#Tuples have limited built-in methods.

t = (1, 2, 2, 3, 4)

print(t.count(2))
print(t.index(3))
#Immutability of Tuples

#You cannot modify elements directly.

t = (1, 2, 3)
# t[0] = 10   # This will cause an error

#If modification is required:

t = (1, 2, 3)
temp = list(t)
temp.append(4)
t = tuple(temp)
print(t)
```

When to Use Tuples

Use tuples when:

Data should not change

You want safer data handling

You are returning multiple values from a function
2. Sets
What is a Set?

A set is:

Unordered

Mutable

Stores only unique elements
Creating Sets

```python
s = {1, 2, 3}
empty_set = set()
#Adding Elements
s.add(4)
s.add(2)  # Duplicate will not be added
#Removing Elements
s.remove(3)   # Error if not found
s.discard(10) # No error if not found
#Other Set Methods
s.pop()
s.clear()
#Removing Duplicates Using Set
lst = [1, 1, 2, 3, 3, 4]
unique = set(lst)
print(unique)
```

3. Dictionaries
What is a Dictionary?

A dictionary stores data in key-value pairs.

Keys must be unique

Values can be any data type

Creating Dictionaries


```python
d = {"name": "John", "age": 25}

print(d["name"])
#Mixed Data Types
d = {
    "numbers": [1, 2, 3],
    "info": {"city": "Dallas"}
}

print(d["numbers"][0])
#Updating Values
d["age"] = 26
d["age"] += 1
#Adding New Keys
d["country"] = "USA"
#Dictionary Methods
d = {"a": 1, "b": 2}

print(d.keys())
print(d.values())
print(d.items())
```

```python
#Nested Dictionaries
d = {"outer": {"inner": {"value": 100}}}

print(d["outer"]["inner"]["value"])
#Dictionary Comprehension
squares = {x: x*x for x in range(1, 6)}
print(squares)
```

7. Generators

Generators produce values one at a time using yield.

They are memory efficient.

Example


```python
def generate_squares(n):
    for i in range(n):
        yield i * i

for num in generate_squares(5):
    print(num)
```

Fibonacci Generator

```python
def fibonacci(n):
    a, b = 1, 1
    for i in range(n):
        yield a
        a, b = b, a + b
print(list(fibonacci(5)))
```

8. map()

Applies a function to every element of an iterable.

```python
nums = [1, 2, 3]

result = list(map(lambda x: x + 1, nums))
print(result)
```

9. reduce()

Reduce repeatedly applies a function and returns a single result.


```python
from functools import reduce

nums = [1, 2, 3, 4]

total = reduce(lambda a, b: a + b, nums)
print(total)
```

10. filter()

Filters elements based on a condition.

```python
nums = [1, 2, 3, 4, 5]

evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
#Shallow vs Deep Copy
#Shallow Copy
import copy

original = [1, 2, 3]
shallow = copy.copy(original)

shallow[0] = 100
print(original)
print(shallow)
#Deep Copy
import copy

original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)

deep[0][0] = 100

print(original)
print(deep)
```
