# For loop examples:

1. Factorial of a Number 

```python
n=int(input("Give input number to caluclate factorial"))
if n<0:
    print("Not a positive number")
else:
    result=1
    for i in range(1,n+1):
        result=result*i
    print(f"result",result)
```

2. Check prime

```python
import math
n=int(input("Enter number to check prime or not"))
if n<=1:
    print("not a prime")
else:
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            print("not a prime")
            break; 
    else:
        print(n,"is prime no")
```

3. Second largest in a list

```python
my_list=list(map(int,input("Enter list of no seperated by space").split()))
print(my_list)
second=largest=float('-inf')
for i in my_list:
    if i>largest:
        second=largest
        largest=i
    elif second<i and i!=largest:
        second=i
if second==float('-inf'):
    print("all no's in list are equal")
else:
    print("second largest no is",second)
```

4.Count Duplicates in a list

```python
new_list=list(map(int,input("Enter numbers seperated by space").split()))
print(new_list)
count={}
for i in new_list:
    if i in count:
        count[i]=count[i]+1
    else:
        count[i]=1

duplicates=False

for key,value in count.items():
    if value>1:
        print(f"{key}:{value}")
        duplicates=True
if duplicates==False:
    print("no duplicates found")

    
```

Count Total Number of Duplicate Elements

```python
nums = list(map(int, input("Enter numbers separated by space: ").split()))

freq = {}
duplicate_count = 0

for n in nums:
    freq[n] = freq.get(n, 0) + 1

for value in freq.values():
    if value > 1:
        duplicate_count += 1

print("Total duplicate elements:", duplicate_count)
```

# While loop Examples

1. Check Palindrome

```python
import math
n=int(input("Enter number to check palindrome"))
if n<1:
    print("not a positive")
else:
    original=n
    rev=0
    while(n>0):
        r=n%10
        rev=rev*10+r
        n=n//10
if original==rev:
    print("palindrome")
else:
    print("not a palindrome")
```

2.Fibonacci Using While

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1
count = 0

while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```

3.Menu-Driven Program

```python
while True:
    print("\n1. Say Hello")
    print("2. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Hello!")
    elif choice == 2:
        print("Exiting...")
        break
    else:
        print("Invalid choice")
```

```python
4.Armstrong Number
```

```python
n = int(input("Enter number: "))
original = n
digits = len(str(n))
total = 0

while n > 0:
    digit = n % 10
    total += digit ** digits
    n //= 10

if total == original:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
5.Reverse a String Using While (No Slicing)
```

```python
s = input("Enter string: ")
i = len(s) - 1
rev = ""

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed string:", rev)
```
