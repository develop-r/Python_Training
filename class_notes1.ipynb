```python
count = 5
while count > 0:
    print("Counting down:", count)
    count -= 1
```

```python
for i in range(10):
    if i == 4:
        break
    print(i)

for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
numbers = [5, 2, 9, 1, 7]

total = 0
for num in numbers:
    total += num
print(total)

rev = []
for num in numbers:
    rev = [num] + rev
print(rev)
```

```python
correct_name = "Alice"

while True:
    user_input = input("Enter your name: ")
    if user_input == correct_name:
        print("Welcome, Alice!")
        break
    else:
        print("Name incorrect. Try again.")
```
